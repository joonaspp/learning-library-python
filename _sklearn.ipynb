# Modelos de Classificação

```python
#carregando base iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#imagem dos target da base iris
#!pip install PIL
from PIL import Image
img = Image.open('images/iris.png')
img
```

```python
#fazendo split da base em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4)
```

## Regressão Logística

É usada para estimar valores discretos (valores binários como 0/1, sim/não, verdadeiro/falso) baseado em um grupo de variáveis independentes. Em palavras simples, ela prevê a probabilidade da ocorrência de um evento, ajustando os dados a uma função logística. Por isso, também é conhecida como regressão logística. Como prevê a probabilidade, seus valores de saída são algo esperado entre 0 e 1.

A classificação multiclasse com regressão logística pode ser feita pelo esquema one-vs-rest, no qual é resolvido para cada classe um problema de classificação binária de dados pertencentes ou não àquela classe ou alterando a função de perda para perda de entropia cruzada.

```python
img = Image.open('images/logistic_regression.png')
img
```

```python
img2 = Image.open('images/logistic_regression_form.png')
img2
```

```python
#criando e treinando modelo de regressão logistica
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
#armazenando os dados preditos em uma variavel
y_pred = logreg.predict(X_test)
```

```python
#avaliando modelo com todas as métricas
print(classification_report(y_test,y_pred))
```

## KNN (K Nearest Neighbor)

O algoritmo de classificação baseado no vizinho mais próximo (Nearest Neighbor - NN) é uma técnica amplamente empregada para reconhecer padrões. O centro de seu funcionamento está em descobrir o vizinho mais próximo de uma dada instância. O algoritmo (k-Nearest Neighbor - k-NN) pertence a um grupo de técnicas denominada de Instance-based Learning. Nesse caso, são encontrados os k vizinhos mais próximos do padrão de consulta, ao invés de apenas o vizinho mais próximo. Dito dessa forma, esse método é bastante simples e de fácil implementação.<br><br>
O algoritmo calcula a distância do elemento dado para cada elemento da base de treinamento e então ordernar os elementos da base de treinamento do mais próximo ao de maior distância. Dos elementos ordenados selecionam-se apenas o k primeiros, que servem de parâmetro para a regra de classificação.

```python
img = Image.open('images/knn.png')
img
```

```python
img2 = Image.open('images/knn_distance.png')
img2
```

```python
#descobrindo a melhor n para o modelo de knn
k_range = list(range(1, 30))
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
    
%matplotlib inline
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

```python
#criando e treinando modelo de knn
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
#armazenando os dados preditos em uma variavel
y_pred = knn.predict(X_test)
```

```python
#avaliando modelo com todas as métricas
print(classification_report(y_test,y_pred))
```

## SVM (Support Vector Machine)

<b>Linear<b>

```python
#criando e treinando modelo de svm linear
model = LinearSVC()
model.fit(X_train, y_train)
```

```python
#armazenando os dados preditos em uma variavel
y_pred = model.predict(X_test)
```

```python
#avaliando modelo com todas as métricas
print(classification_report(y_test,y_pred))
```

<b>Poly<b>

```python
#criando e treinando modelo de svm poly
model = SVC(kernel='poly', degree=8)
model.fit(X_train, y_train)
```

```python
#armazenando os dados preditos em uma variavel
y_pred = model.predict(X_test)
```

```python
#avaliando modelo com todas as métricas
print(classification_report(y_test,y_pred))
```

<b>Sigmoid<b>

```python
#criando e treinando modelo de svm sigmoid
model = SVC(kernel='sigmoid')
model.fit(X_train, y_train)
```

```python
#armazenando os dados preditos em uma variavel
y_pred = model.predict(X_test)
```

```python
#acurácia
print(accuracy_score(y_test,y_pred))
```

```python
#avaliando modelo com todas as métricas
print(classification_report(y_test,y_pred))
```

<b>Gaussian Kbf<b>

```python
#criando e treinando modelo de svm gaussian kbf
model = SVC(kernel='rbf')
model.fit(X_train, y_train)
```

```python
#armazenando os dados preditos em uma variavel
y_pred = model.predict(X_test)
```

```python
#avaliando modelo com todas as métricas
print(classification_report(y_test,y_pred))
```
